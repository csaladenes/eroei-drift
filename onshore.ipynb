```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from jupyterthemes import jtplot
# oceans16 | grade3 | chesterish | onedork | monokai | solarizedl
jtplot.style('solarized-light')
```

```python
df=pd.read_excel('data/wind_farms_with_ws_2k_5k_sgouris edited.xlsx',sheetname='BASE')
df=df.dropna()
df=df.set_index(['ISO_CODE','Year'])
df=df[((df['POWER_KW']>0)&(df['OFFSHORE']=='no'))]
```

```python
cs={'BR','AR','GB','CA','US','IN'}
for c in cs:
    x=[]
    y=[]
    z=[]
    for i in df.loc[c].sort_index().index.unique():
        x.append(i)
        #y.append(np.median(df.loc[c].loc[i]['WS_est']))
        z.append(np.mean(df.loc[c].loc[i]['WS_est']))
    #plt.plot(x,y,label=c)
    plt.plot(x,z,label=c)
plt.ylabel('Mean windspeed [m/s]')
plt.title("Onshore turbines")
plt.legend(loc=0)
```

```python
df=pd.read_excel('data/wind_farms_with_ws_2k_5k_sgouris edited.xlsx',sheetname='BASE')
df=df.dropna()
df=df.set_index('Year')
df=df[((df['POWER_KW']>0)&(df['OFFSHORE']=='no'))]
```

```python
x=[]
y=[]
z=[]
for i in df.sort_index().index.unique():
    x.append(i)
    y.append(np.median(df.loc[i]['WS_est']))
    z.append(np.mean(df.loc[i]['WS_est']))
plt.plot(x,y,label='WD median')
plt.plot(x,z,label='WD mean')
plt.ylabel('Mean windspeed [m/s]')
plt.title("Onshore turbines")
plt.legend(loc=0)
```
